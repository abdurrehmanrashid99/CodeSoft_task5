```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression

# Load the credit card dataset
credit_card_data = pd.read_csv('C:\\Users\\ms\\PycharmProjects\\pythonProject1\\creditcard.csv')
# Explore the class distribution
class_counts = credit_card_data['Class'].value_counts()
normal_percentage = round(class_counts[0] / len(credit_card_data) * 100, 2)
fraud_percentage = round(class_counts[1] / len(credit_card_data) * 100, 2)
print("Percentage of normal transactions:", normal_percentage)
print("Percentage of fraud transactions:", fraud_percentage)

# Calculate the correlation matrix
correlation_matrix = credit_card_data.corr()

# Visualize the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True)
plt.show()

# Separate legitimate and fraudulent transactions
legitimate_transactions = credit_card_data[credit_card_data['Class'] == 0]
fraudulent_transactions = credit_card_data[credit_card_data['Class'] == 1]

# Explore statistics of legitimate transactions
legitimate_amount_stats = legitimate_transactions['Amount'].describe()

# Sample an equal number of legitimate transactions to balance the dataset
sampled_legitimate = legitimate_transactions.sample(n=len(fraudulent_transactions))

# Combine the sampled legitimate and fraudulent transactions
balanced_dataset = pd.concat([sampled_legitimate, fraudulent_transactions], axis=0)

# Create a derived column for hours
balanced_dataset['Hour'] = (balanced_dataset['Time'] / 3600).astype(int)

# Drop the 'Time' column
balanced_dataset.drop(columns='Time', inplace=True)

# Define features (X) and target (y)
X = balanced_dataset.drop('Class', axis=1)
y = balanced_dataset['Class']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=3, stratify=y)

# Visualize the distribution of features for fraud and non-fraud cases
plt.figure(figsize=(12, 40))
for i, col in enumerate(X.columns):
    plt.subplot(14, 3, i + 1)
    sns.histplot(X[col][y == 0], color='blue', kde=True, stat='density', label='Non-Fraud', alpha=0.6)
    sns.histplot(X[col][y == 1], color='red', kde=True, stat='density', label='Fraud', alpha=0.6)
    plt.title(col, fontsize=14)
    plt.legend()
plt.tight_layout()
plt.show()

# Train a logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Predict on the training set
y_train_pred = model.predict(X_train)

# Predict on the test set
y_test_pred = model.predict(X_test)

# Calculate and display accuracy scores
train_accuracy = round(accuracy_score(y_train_pred, y_train) * 100, 2)
test_accuracy = round(accuracy_score(y_test_pred, y_test) * 100, 2)
print('Training accuracy:', train_accuracy)
print('Testing accuracy:', test_accuracy)

# Generate a classification report for the test set
class_report = classification_report(y_test, y_test_pred)
print('Classification report for the test set:\n', class_report)
```
